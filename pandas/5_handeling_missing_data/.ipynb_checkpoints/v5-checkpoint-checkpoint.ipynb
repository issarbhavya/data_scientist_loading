```python
"""
since the date is in string format, we need to convert it into :
Timestamp,  this can be done by 
"""
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=["day"])

## now check the type of date
type(df.day[1])


## syntax : parse_dates=[column name]
```

```python
## setting dates as index to make handeling easier

df.set_index("day") ### just doing this will not save day as index,
## for that, we do inplace=True

df.set_index("day",inplace=True)
```

```python
df
```

```python
"""
if you want to add same data to all na, then you can use

new_df= df.fillna(0)
new_df

"""

new_df=df.fillna({
    "temperature":0,
    "windspeed":0,
    "event":"no event"
})

new_df
```

```python
 df
```

```python
"""
the value 0 isn't a better data, instead u can use *method* func
"""

new_df=df.fillna(method="ffill")
new_df

## puts the previous data in the na cell
```

```python
new_df=df.fillna(method="bfill")
new_df

##puts next data in previous missing one 
```

```python
new_df=df.fillna(method="bfill", axis="columns")
new_df

## now the method's axis is changed, it will read columns by columns
```

```python
df
```

```python
"""
using interpolate to fill the na values
this will give a mean value of the cels above nd below
to make data more precise 
"""

new_df=df.interpolate()
new_df
```

```python
"""
if you focus on date then you see that the dates are 1st then 4th,
2nd nd 3rd are missing so using interpolate won't give data precisely
therefore, use   *method=Time*
"""

new_df=df.interpolate(method="time")
new_df
```

```python
## **************

#### dropna()

## **************


newdf=df.dropna()
newdf
```

```python
newdf=df.dropna(how="all")
newdf

### for the date 9th, all the values of the rows were na, and the date was inde, soother then index,
### all was empty so that row was removed
```

```python
new_df = df.dropna(thresh=2)
new_df

### this means that if there are two valid
### values then the row might not be excluded
```

```python
### to insert the mising dates/rows, first make a date range 
### then we pass that to date time index 
### then do reindexing nd either update it to the existing or,
### save in a new dataframe

dt=pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
new_df = df.reindex(idx)
new_df


### now you can use of the filling methods to fill the na
```

```python
new_df_2=new_df.interpolate(method="time")
new_df_2
```
