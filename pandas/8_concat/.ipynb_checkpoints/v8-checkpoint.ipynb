```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
### 2 data frames are made

df = pd.concat([india_weather, us_weather])
df
```

```python
## in the one above, you see index as 0,1,2 nd then again 0,1,2

## to solve this

df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
## TO make keys, showing the concatenated data at the same time
## implementing that the tables are differenciable 

df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
## throgh the key function, u can still call the tables seprately

df.loc["us"]
```

```python
####


# NOW iF YOU WANT TO ADD DATA COLUMNS


####

df = pd.concat([india_weather, us_weather],axis=1)
df
```

```python
####

# Concatenation Using Index

####
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])  
windspeed_df

## giving similar index to cities so that data dosen't mismatch

## done by specifying index at the time of dataframe creation
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

```python
#### you can concat a series as well

# create a series that will work as a row or column as per ur wish
# nd assign data to itin form of list 

s = pd.Series(["Humid","Dry","Rain"], name="event")
s
```

```python
new_df = pd.concat([temperature_df,s],axis=1)
new_df

# axis=1 as i want to add a column
```

```python
## how to cancat a row, through serie method

s2 = pd.Series(["kolkata","33","rain"], name="row3")
s2
```

```python
new_df_2 = pd.concat([new_df,s2],axis=1)
new_df_2
```
