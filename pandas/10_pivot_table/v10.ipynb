```python
##

# PIVOT

##

"""
"pivot" requires 2 arguments to alter the table, one is index,
other is columns

index is used to specify the selection of rows, whereas
columns specfies columns
""" 

df2=df.pivot(index="date",columns="city")
df2
```

```python
### if you want specific values in the table use "values

df2=df.pivot(index="date",columns="city",values="temperature")
df2
```

```python
##

## pivot_table

##

##used to summarize nd aggregate data inside df

df=pd.read_csv("weather2.csv")
df
```

```python
## as you can see that data has 2 temperature for same dates,thus
## on pivoting, it puts mean of humidity nd temp for each date

df.pivot_table(index="city",columns="date")
```

```python
## aggfunc="" is used take desired actions on datas belonging
## to same index, aggfunc="mean" is default

df.pivot_table(index="city",columns="date",aggfunc="sum")

## using sum returns the sum of the 2 values of temp nd humidity
## for the same date
```

```python
## similarly, count will show the number of data avaliable for
## same date

df.pivot_table(index="city",columns="date",aggfunc="count")
```

```python
## margins gives mean on all dataset ends

df.pivot_table(index="city",columns="date",margins=True)
```

```python
## you can change its name as well

df.pivot_table(index="city",columns="date",margins=True,margins_name="avg")
```

```python
##

# DIVIDING DATA ON BASIS OF DATES, CAN BE MONTHS,YEARS,WEEKS,ETC

##

df = pd.read_csv("weather3.csv")
df
```

```python
## converting date which was in string to datetime 

df = pd.read_csv("weather3.csv",parse_dates=["date"])
df
```

```python
## another way of converting str to datetime

df['date'] = pd.to_datetime(df['date'])
type(df.date[0])
```

```python
## now we can use grouper to divide the data as per
## the date to our accordance

df.pivot_table(index=pd.Grouper(freq="M",key="date"),columns="city")
```
