```python
import pandas as pd 
import numpy as np
df=pd.read_csv("weather_data2.csv")
df
```

```python
newdf=df.replace(-99999, np.NaN)
newdf
```

```python
newdf=df.replace([-99999,-88888],np.NaN)
newdf
```

```python
newdf=df.replace("no event",np.NaN)
newdf
```

```python
### replacing specific values IN A COLUMN using dictionary
### COLUMN BY COLUMN SPECIFICATION

### dis: a column might contain many waste values but we have to
### seprately specify all
    
    
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -88888,
        'event': 'no event'
    }, np.NaN)
new_df
```

```python
### replacing specific values IN ENTIRE SHEET using dictionary
###  FOR COMPLETE SHEET use *** replace method by mapping ***

new_df = df.replace({
        -99999: np.NaN,
        -88888: np.NaN,
        'no event': 'Sunny',
    })
new_df
```
